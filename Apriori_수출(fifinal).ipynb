```python
def old_groups(values):
    if values > 0:
        return 1
    else:
        return 0
```

```python
SampleData2['cylinderRate'] = SampleData2['cylinderRate'].apply(old_groups)
SampleData2['memoryRate'] = SampleData2['memoryRate'].apply(old_groups)
SampleData2['machine_componentRate'] = SampleData2['machine_componentRate'].apply(old_groups)
SampleData2['printed_circuit_boardRate'] = SampleData2['printed_circuit_boardRate'].apply(old_groups)
SampleData2['solid_stateRate'] = SampleData2['solid_stateRate'].apply(old_groups)
SampleData2['processor_controllerRate'] = SampleData2['processor_controllerRate'].apply(old_groups)

SampleData2.head
```

```python
frequent_itemset = apriori(SampleData2,
                           min_support=0.005, 
                           max_len=3, 
                           use_colnames=True, 
                           verbose=1 
                          )
frequent_itemset['length'] = frequent_itemset['itemsets'].map(lambda x: len(x))
frequent_itemset.sort_values('support',ascending=False,inplace=True)
frequent_itemset
```

```python
frequent_itemset.loc[frequent_itemset['length']> 1]
```
