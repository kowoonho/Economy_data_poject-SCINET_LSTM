```python
from scipy import signal
def ccf_values(series1, series2):
    p = series1
    q = series2
    p = (p - np.mean(p)) / (np.std(p) * len(p))
    q = (q - np.mean(q)) / (np.std(q))  
    c = np.correlate(p, q, 'full')
    return c
    
```

```python
ccf_ielts1 = ccf_values(df['산업생산지수(원지수)'], df['생산자제품 재고지수(원지수)'])
a = np.max(ccf_ielts1)
a
```
